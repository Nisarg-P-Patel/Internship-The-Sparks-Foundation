```python
##importing necessary libraries

import numpy
import pandas as pd
from sklearn import datasets as dt
from sklearn import metrics as mt 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#fetching data from dataset

X,Y=dt.load_iris(return_X_y=True)

iris_data=dt.load_iris()
```

```python
#printing first 5 entries of dataset

iris_df=pd.DataFrame(iris_data.data,columns=iris_data.feature_names)
iris_df=iris_df.sample(frac=1)
iris_df.head()
```

```python
#describing all details of the data

print(iris_df.shape)
iris_df.describe()
```

```python
#plotting the data according to features "SEPAL"

x=iris_df["sepal length (cm)"]
y=iris_df["sepal width (cm)"]
plt.scatter(x,y,color='green')
plt.title('sepal width (cm) Vs sepal length (cm)')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
#plotting the data according to features "PETAL"

x=iris_df["petal length (cm)"]
y=iris_df["petal width (cm)"]
plt.scatter(x,y,color='green')
plt.title('petal width (cm) Vs petal length (cm)')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```

```python
#elbow method 

inertia_value=[]

for i in range(1, 11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter = 300, n_init = 10, random_state = 0).fit(X)
    inertia_value.append(kmeans.inertia_)
    
# Plotting the results onto a line graph
plt.plot(range(1,11),inertia_value,color='green',marker='o',markerfacecolor='pink')
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia Value')
```

```python
# Applying kmeans with k=3 to the dataset 

kmeans=KMeans(n_clusters=3,init='k-means++',max_iter = 300, n_init = 10, random_state = 0).fit(X)
y_kmeans=kmeans.fit_predict(X)
```

```python
# Visualising the predicted cluster

# Visualising the clusters - On the first two columns

plt.figure(figsize=(12,8))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1],alpha=.75, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1],alpha=.75, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1],alpha=.75, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s = 75, c = 'orange', label = 'Centroids')
plt.legend(loc='upper right',fontsize=15)
```

```python
# Visualising the real clustering of dataset

# Visualising the clusters - On the first two columns

plt.figure(figsize=(12,8))
plt.scatter(X[Y == 0, 0], X[Y == 0, 1],alpha=.75, c = 'red', label = 'Iris-setosa')
plt.scatter(X[Y == 1, 0], X[Y == 1, 1],alpha=.75, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[Y == 2, 0], X[Y == 2, 1],alpha=.75, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s = 75, c = 'orange', label = 'Centroids')
plt.legend(loc='upper right',fontsize=15)
```
